# Bird Strikes Analysis Between 2000-2011

## Problem Statement 

- Transport and communication are in the crucial domain in the field of analytics. Environmental impacts and safety are, nowadays, two major concerns of the scientific community with respect to transport scenarios and to the ever-growing urban areas. These issues gain more importance due to the increasing amount of vehicles and people. Seeking new solutions is reaching a point where available technologies and artificial intelligence, especially MAS, are being recognized as ways to cope with and tackle these kinds of problems in a distributed and more appropriate way.
 
 
- A bird strike is strictly defined as a collision between a bird and an aircraft which is in flight or on a take-off or landing roll. The term is often expanded to cover other wildlife strikes - with bats or ground animals. Bird Strike is common and can be a significant threat to aircraft safety. For smaller aircraft, significant damage may be caused to the aircraft structure and all aircraft, especially jet-engine ones, are vulnerable to the loss of thrust which can follow the ingestion of birds into engine air intakes. This has resulted in several fatal accidents.
 
 
- Bird strikes may occur during any phase of flight, but are most likely during the take-off, initial climb, approach and landing phases due to the greater numbers of birds in flight at lower levels. To have a closer look the following document visually depicts the data collected on Bird Strikes by FAA between 2000-2011.

## Checking Duplicates and Null Values 

```python
data.describe()
```

```python
# filling the Na value with 'No Remarks' 
data.fillna({'Remarks': 'No Remarks'},inplace=True) 

# filling the Na value with 'Unknown'
data.fillna({'Origin State':'Unknown'},inplace=True) 

# used dictionary to fill the Na value with 'Unknown'
data.fillna({'Airport: Name':'Unknown'},inplace=True) 

# filling the Na value with 'Nothing' 
data.fillna({'Effect: Impact to flight': 'Nothing'}, inplace=True) 

# filling the Na value with 'Unknown'
data.fillna({'Aircraft: Airline/Operator':'Unknown'},inplace=True) 
```

```python
# Used for loop to fill the Na values with the Mode value
for column in ['Aircraft: Type', 'Altitude bin',
                'When: Phase of flight','Aircraft: Number of engines?',
               'Pilot warned of birds or wildlife?', 'Is Aircraft Large?', 'Feet above ground']:
    mode_value = data[column].mode()[0]
    
data[column] = data[column].fillna(mode_value)
```

```python
# Droping the remaining rows with most of the  missing data.
data.dropna(subset=['Wildlife: Number struck','Aircraft: Number of engines?','FlightDate'], inplace=True) 
```

## Yearly Analysis of Bird Strikes in U.S. 

```python
#Including all the relevant states from the 'Origin State' column indicating the state of occurence
bird_strikes_us = data[data['Origin State'].isin(['New York', 'Texas', 'Louisiana', 'Washington', 'Virginia',
       'Unknown', 'Delaware', 'DC', 'Georgia', 'Florida', 'California',
       'Illinois', 'Connecticut', 'Missouri', 'Rhode Island', 'Hawaii',
       'Arizona', 'Tennessee', 'South Carolina', 'South Dakota',
       'New Jersey', 'Colorado', 'Minnesota', 'Alabama', 'Ohio',
       'Wisconsin', 'Michigan', 'Massachusetts', 'Alaska',
       'North Carolina', 'Kentucky', 'Indiana', 'Oregon', 'Pennsylvania',
       'New Hampshire', 'Arkansas', 'Nevada', 'Mississippi', 'Maryland',
       'Maine', 'Quebec', 'Idaho', 'British Columbia', 'Utah', 'Nebraska',
       'Iowa', 'New Mexico', 'West Virginia', 'Oklahoma', 'North Dakota',
       'Vermont', 'Wyoming', 'Kansas', 'Prince Edward Island', 'Montana',
       'Puerto Rico', 'Ontario', 'Virgin Islands',
       'Newfoundland and Labrador', 'Alberta', 'Saskatchewan'])] 

# Group by year of 'Flight date' and count the number of bird strikes
yearly_bird_strikes_in_us = bird_strikes_us.groupby(bird_strikes_us['FlightDate'].dt.year).size()

# Plot the yearly analysis of bird strikes in the US
plt.figure(figsize=(14,8))
bars = yearly_bird_strikes_in_us.plot(kind='bar', color='lightgreen')
plt.title('Yearly Analysis of Bird Strikes in the US')
plt.xlabel('Year')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding labels to each bar
for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), 
                   ha='center', 
                   va='center', 
                   xytext=(0, 5), 
                   textcoords='offset points')

plt.show()
```

- From the above graph, Yearly analysis of bird strikes has been shown from which year '2009' showing the highest count of Bird 
strikes i.e., '3206' following with the year '2010' & '2011'.

## Top 10 US Airlines by Bird Strikes 

```python
# Group by airline/operator and count the number of bird strikes
top_10_airlines = data['Aircraft: Airline/Operator'].value_counts().head(10)

# Plot the top 10 US airlines with the most bird strikes
plt.figure(figsize=(14, 8))  # Set figure size
bars = top_10_airlines.plot(kind='bar', color='violet') 
plt.title('Top 10 US Airlines by Bird Strikes') 
plt.xlabel('Airline/Operator')  
plt.ylabel('Number of Bird Strikes')  
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)  


# Adding labels to each bar
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2,  # x-coordinate
             bar.get_height(),  # y-coordinate
             f'{int(bar.get_height())}',  # Text to display (convert to integer)
             ha='center',  # Horizontal alignment
             va='bottom')  # Vertical alignment

plt.tight_layout() 
plt.show()
```

## Top 10 Airports by Bird Strike 

```python
# Group data by airport name and sum up the number of bird strikes
airport_strikes = data.groupby('Airport: Name')['Wildlife: Number Struck Actual'].sum()

# Plotting
plt.figure(figsize=(14, 8))
top_10_airports = airport_strikes.sort_values(ascending=False).head(10)
top_10_airports.plot(kind='bar', color='skyblue')

# Adding titles and labels
plt.title('Top 10 Airports by Bird Strikes')
plt.xlabel('Airport')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)

# Adding values on top of bars
for i, count in enumerate(top_10_airports):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

- From the above graph, the highest number of bird strucks happens at 'DALLAS/FORT WORTH INTL ARPT' with a count of '2932' following with other Airports with their bird struck count.

## Yearly Cost Incurred due to Bird Strikes 

```python
# Group by year and sum up the total cost for each year
yearly_cost = data.groupby(data['FlightDate'].dt.year)['Cost: Total $'].sum()

# Plot the yearly cost incurred due to bird strikes
plt.figure(figsize=(14, 8))
bars = yearly_cost.plot(kind='bar', color='lightgreen')
plt.title('Yearly Cost Incurred due to Bird Strikes')
plt.xlabel('Year')
plt.ylabel('Cost ($)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding labels to each bar
for bar in bars.patches:
    plt.annotate(format(bar.get_height(), '.2f'), 
                 (bar.get_x() + bar.get_width() / 2, 
                  bar.get_height()), 
                 ha='center', 
                 va='center', 
                 xytext=(0, 5), 
                 textcoords='offset points')

plt.tight_layout()
plt.show()
```

- From the above graph, the maximum cost incurred in the year '2001' following with year '2006' i.e., "23162777.00" & "18302373.00" respectively.

## Bird Strikes by Condition:Precipitations 

```python
data['Conditions: Precipitation'].unique()
```

```python
cond_precip = data.groupby('Conditions: Precipitation')['Wildlife: Number Struck Actual'].sum().to_frame().reset_index().drop(index=4,axis=0)
cond_precip
```

```python
plt.figure(figsize=(15,10))
sns.set(rc={'figure.facecolor':'#F0F3F4'})

# line chart plotting:
plt.subplot(2,2,(3,4))
sns.lineplot(data=cond_precip,x='Conditions: Precipitation',
             y='Wildlife: Number Struck Actual',
             marker='o',mfc='#FF8F00',
             mec='k',ms=23,
             linestyle='--',color='#2E7D32')

# Adding value labels on top of each bar
for i,height in enumerate(cond_precip['Wildlife: Number Struck Actual']):
    plt.text(i,height+(0.05 * height) ,str(height),ha='center',
             va='center',fontweight='bold')
    
plt.xlabel('Weather Conditions',fontdict={'fontweight':'bold','color':'k'})
plt.ylabel('No. of Birds Strike',fontdict={'fontweight':'bold','color':'k'})
plt.title('Birds Strike by Precipitation', fontweight='bold', fontsize=12,color='#B7950B', loc='center') 
```

- The graph shows the total no. of birds struck by different weather conditions, the highest no. of birds struck was when the weather condition was "Rain" with the total of "4374" and lowest when "Fog,Snow" with the total of "4".

## Altitude of aeroplanes at the time of strike 

```python
plt.figure(figsize=(14, 8))
n, bins, patches = plt.hist(data['Altitude bin'], bins=5, color='skyblue', edgecolor='black')
plt.title('Altitude of Airplanes at the Time of Bird Strike')
plt.xlabel('Altitude (feet)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding frequency labels above each bar
for frequency, bin_value, patch in zip(n, bins, patches):
    plt.text(bin_value + (bins[1] - bins[0]) / 2, frequency + 5, str(int(frequency)), ha='center', va='bottom')

plt.show()
```

## Bird Strikes by Phase of Flight 

```python
# Assuming we have a column 'Phase of Flight' indicating the phase of flight at the time of the strike
phase_flight_counts = data['When: Phase of flight'].value_counts()

# Plotting the distribution of bird strikes across different phases of flight
plt.figure(figsize=(14, 8))
phase_flight_counts.plot(kind='bar', color='lightpink')
plt.title('Bird Strikes by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.tight_layout()
plt.show()
```

##  Effect of Bird Strikes & Impact on Flight

```python
Effect_data = data.groupby('Effect: Impact to flight')['Record ID'].count().to_frame().reset_index()
Effect_data
```

```python
plt.figure(figsize=(25,10))

sns.set(rc={'figure.facecolor':'#B2EBF2'})

colors=['#FBC02D','#D32F2F','#239B56','#4FC3F7','#000099']
explodes=[0.3,0.4,0.1,0.1,0.2]

plt.pie(Effect_data['Record ID'],
        labels=Effect_data['Effect: Impact to flight'],
        radius=1,
        colors=colors,
        explode=explodes,
        shadow=True,labeldistance=1.05,
        textprops={'fontsize':10,'fontweight':'bold','color':'#000000'},
        wedgeprops={'linewidth':1,'edgecolor':'#000000'})

plt.pie([1],radius=0.5,colors=['#B2EBF2'],
        wedgeprops={'linewidth':1.5,'edgecolor':'#000000'}) # second pie chart to make donot

plt.title('Distribution of Impact to Flight Caused by Birds Strike', 
          fontweight='bold',fontsize=15,color='#FF6F00',loc='right')

plt.show()
```

## Effect of Strikes at Different Altitude 

```python
Altitude_data = data.groupby('Altitude bin')['Wildlife: Number Struck Actual'].sum().to_frame().reset_index()
Altitude_data
```

```python
plt.figure(figsize=(14, 8))

bar= sns.barplot(x='Altitude bin', y='Wildlife: Number Struck Actual', 
                 data=Altitude_data, color='#40E0D0', edgecolor='black',lw=1.5)

# Adding value labels on top of each bar

for i,height in enumerate(Altitude_data['Wildlife: Number Struck Actual']):
    plt.text(i,height -(0.07*height),str(height),ha='center',
             va='top',fontweight='bold',color='#6433FF')

plt.xlabel('Altitude bin', fontweight='bold')
plt.ylabel('Number of Wildlife Strikes', fontweight='bold')
plt.title('Effect of Wildlife Strikes By Altitude', 
          fontweight='bold',color='#F80101',fontsize=20)

bar.tick_params(axis='x',colors='#17208A') 
bar.tick_params(axis='y',colors='#17208A') 

plt.show()
```

##  Prior Warning and Effect of Strike Relation 

```python
# Filter out NaN values in 'Effect: Impact to flight' and 'Pilot Warned' columns
data_cleaned = data.dropna(subset=['Effect: Impact to flight', 'Pilot warned of birds or wildlife?'])

# Set up the plot
plt.figure(figsize=(14, 8))

# Create a count plot
ax = sns.countplot(x='Effect: Impact to flight', hue='Pilot warned of birds or wildlife?', data=data_cleaned, palette='viridis')

# Add labels and title
plt.title('Prior Warning and Effect of Bird Strike')
plt.xlabel('Effect on Flight')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)  

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 0.1, height, ha='center')

plt.tight_layout()
plt.show()
```
